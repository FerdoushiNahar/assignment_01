```python
import pandas as pd
import numpy as np
from matplotlib import pyplot as plt
```

```python
df=pd.read_csv('Height_weight_Data.csv')
```

```python
x = df[['Height']]
y = df['Weight']
```

```python
from sklearn.model_selection import train_test_split as tts
```

```python
xtrain,xtest,ytrain,ytest=tts(x,y,test_size=.30)
```

```python
from sklearn.linear_model import LinearRegression
```

```python
reg = LinearRegression()
```

```python
reg.fit(xtrain, ytrain) 
```

```python
pred = reg.predict(xtest)
```

```python
pred
```

```python
df['predicted result'] = reg.predict(x)
```

```python
from sklearn.metrics import mean_squared_error, mean_absolute_error
```

```python
mse = mean_squared_error(ytest, pred)
```

```python
mse
```

```python
mae = mean_absolute_error(ytest, pred)
```

```python
mae
```

```python
plt.figure(figsize=(12,8))
plt.scatter(df.Height , df.Weight, marker='^' , color='#cb357b')
plt.xlabel('Height', color='black')
plt.ylabel('Weight')
plt.title('Weight Prediction', color='green')
plt.plot(xtrain , reg.predict(xtrain), color='green')
```
